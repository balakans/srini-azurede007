#### FIlter vs Select API
- filter - filter the rows
- select - filter the columns

```python
input_path = "/Volumes/inceptez_catalog/inputdb/customerdata/custs"

### Read CSV
df = (
    spark.read.format("csv")
    .option("inferSchema", "true")
    .load(input_path)
    .toDF("custid","fname","lname","age","profession")
)
df.display()
```

```python
#select * from cust where profession = 'Pilot'
df.filter(df.profession == 'Pilot').display()

#select custid, age, profession from cust
df.select("custid","age","profession").display()
```
