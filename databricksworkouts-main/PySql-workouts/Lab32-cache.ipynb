```python
df = spark.table("inceptez_catalog.inputdb.tbltxns")
df1 = df.filter("state = 'California'")
df1.cache()
#or
df1.persist()

from pyspark.sql import storagelevel

#storage level

df1.persist(storagelevel.DISK_ONLY)
df1.persist(storagelevel.MEMORY_ONLY)
df1.persist(storagelevel.DISK_ONLY_2)
df1.persist(storagelevel.MEMORY_ONLY_SER)
df1.persist(storagelevel.MEMORY_AND_DISK)

#unpersist() - To remove the data from cache

df1.unpersist()






df1.display()
```
