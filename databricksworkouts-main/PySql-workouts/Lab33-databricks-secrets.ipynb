### Use Secrets in Notebook using dbutils

```python
# List the scope in the databricks account
display(dbutils.secrets.listScopes())
```

```python
# List the secrets under the scope
display(dbutils.secrets.list("insurance_scope"))
```

```python
userid = dbutils.secrets.get(scope="insurance_scope", key="mysql_userid")
print(userid)
pwd = dbutils.secrets.get(scope="insurance_scope", key="mysql_password")
print(pwd)

jdbc_url = f"jdbc:mysql://insurance.mysql.database.azure.com:3306/insurancedb&user={userid}&password={pwd}"
# Read the data from mysql database
insurance_df = spark.read.format("jdbc").option("url", jdbc_url).option("dbtable", "tblcustomer").load()
```
